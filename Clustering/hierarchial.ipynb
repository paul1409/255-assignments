Plot arbitrary data points

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage

x1 = np.array([1,1,2,2,3,3,3,3,4,4,5,5,6,7,7,7,8,8])
x2 = np.array([5,7,6,7,3,6,8,9,5,8,4,7,9,2,3,9,1,7])
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)

import matplotlib.pyplot as plt
labels = range(1, 19) 
plt.figure(figsize=(10, 10)) 
plt.subplots_adjust(bottom=0.1) 
plt.scatter(X[:,0],X[:,1], label='True Position')
for label, x, y in zip(labels, X[:, 0], X[:, 1]): 
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

Use a dendogram to get estimate of clusters in single link

```python
linked = linkage(X, 'single')
labelList = range(1, 19)
plt.figure(figsize=(10, 10)) 
dendrogram(linked, 
    orientation='top',
    labels=labelList,
    distance_sort='descending',
    show_leaf_counts=True)
plt.show()
```

Plot clusters

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='single')
print(cluster.fit_predict(X))
plt.figure(figsize=(10, 10))
plt.scatter(X[:,0], X[:,1], c=cluster.labels_, cmap='rainbow') 

plt.show() 
```
